```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
simulator = AerSimulator()

qc = QuantumCircuit(1, 1)

# Put qubit in superposition
qc.h(0)

# Measure
qc.measure(0, 0)

# Run simulation
job = simulator.run(qc, shots=1000)

result = job.result()

counts = result.get_counts()

print("Quantum Random Numbers")
print(counts)
```

```python
plot_histogram(counts)
```

```python
zeros = counts.get("0", 0)

ones = counts.get("1", 0)

print("Zeros:", zeros)

print("Ones:", ones)

print("Mean:", np.mean([zeros, ones]))

print("Standard Deviation:", np.std([zeros, ones]))
```

```python
classical = [random.randint(0,1) for _ in range(20)]

print("Classical Random Numbers")

print(classical)
```

```python
print("COMPARISON")

print("\nClassical Randomness")
print("- Generated using algorithm")
print("- Pseudo-random")
print("- Deterministic with seed")

print("\nQuantum Randomness")
print("- Generated using quantum superposition")
print("- Measurement based")
print("- Intrinsically probabilistic")
```
